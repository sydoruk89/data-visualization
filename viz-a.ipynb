```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('./datasets_748584_1294786_2015.csv')
df.head()
```

```python
# Linear Regression Happiness Score and Economy
sns.set()
plt.figure(figsize=(8, 5), dpi=300)
sns.regplot(x="Economy (GDP per Capita)", y="Happiness Score", data=df).set(title='Happiness Score and Economy(GDP)')
```

```python
region = df.groupby('Region', as_index=False)
central_eu = region.get_group('Central and Eastern Europe')
```

```python
# Central and Eastern EU
sns.set()
sns.relplot(x="Economy (GDP per Capita)", y="Happiness Score", hue='=Country', palette='rocket', size="Happiness Score", sizes=(70, 350), data=central_eu).set(title='Happiness Score and Economy(GDP)')
```

```python
#freedom score and Trust Government (mean by region)
mean = region.mean()
sns.relplot(x="Freedom", y="Trust (Government Corruption)", size="Happiness Score", sizes=(70, 350), palette='rainbow', hue="Region", data=mean)
```

```python
w_eu = region.get_group('Western Europe')
```

```python
f, ax = plt.subplots(figsize=(7, 15))
sns.set_color_codes("muted")
sns.barplot(x="Happiness Score", y="=Country", data=w_eu, color="b")
```
