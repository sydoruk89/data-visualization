```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('./migration_nz.csv')
df.head()
```

```python
year = df.groupby('Year', as_index=False).sum()
f, ax = plt.subplots(figsize=(15, 8))
g = sns.barplot(x="Year", y="Value", data=year, palette="Blues_d")
y = sorted(df['Year'].value_counts().index.tolist())
g.set_xticklabels(labels=y, rotation=45)
g.set(title='Migration by year')
```

```python
country = df.groupby('Country', as_index=False).sum().sort_values(by='Value', ascending=False)
```

```python
top_20_with_all_row = country.head(21)
top_20 = top_20_with_all_row[1:]
```

```python
f, ax = plt.subplots(figsize=(7, 15))
sns.set_color_codes("muted")
sns.barplot(x="Value", y="Country", data=top_20, color="b").set(title='Top 20 country migration')
sns.set_color_codes("pastel")
```

```python
sns.relplot(x="Year", y="Value", size="Value", sizes=(75, 300), data=year).set(title='Migration by year')
```
